```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss='mean_squared_error')
#Optimazer es como el error permitido y al mismo tiempo es la tasa de cambio del modelo.
#Entre más preciso puede ser mas tardado el modelo o llegar mas rapido al resultado esperado.
```

```python
print("Entrenamiento")
historial = modelo.fit(celsius, fahrenheit, epochs=100, verbose=False)
print("Terminado")
#Epoch son los ciclos de entrenamiento, se pueden ajustar despues de una primera corrida
#con el objetivo de reducir ciclo inecesario ya que no existe cambios significativos en el modelo.
```

```python
import matplotlib.pyplot as plt
plt.xlabel("#Ciclo")
plt.ylabel("Magnitud de Perdida")
plt.plot(historial.history["loss"])
```

```python
print("Predicción")
predi = modelo.predict([100])
print("El resultado es " + str(predi) + "Fahrenheit")
```
